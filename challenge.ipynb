  ## Identifying Outliers using Standard Deviation

```python
# initial imports
import pandas as pd
import numpy as np
import random
from sqlalchemy import create_engine

```

```python
# create a connection to the database
engine = create_engine("postgresql://postgres:*789Wowwe@localhost:5432/sql_homework")
query= """ SELECT * FROM customer_spending;"""
```

```python
# code a function to identify outliers based on standard deviation
spending_df = pd.read_sql(query, engine)
spending_df.sort_values(by=['amount'], ascending=False)
```

```python
std_df = spending_df.groupby("id_chr").apply(np.std)
std_df = std_df.drop(columns= ['id_chr'])
std_df = std_df.sort_values(by=['amount'], ascending=False)
std_df.head()
```

```python
explore_df= spending_df.groupby("id_chr")
explore_df.sort_values("id_chr", ascending=True)
```
